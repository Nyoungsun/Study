```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

# 데이터

```python
dataset = np.array(range(1, 11)) 
timesteps = 5

def split_x(dataset, timesteps):
    tmp = []
    for i in range(len(dataset) - timesteps + 1):
        subset = dataset[i : (i + timesteps)]
        tmp.append(subset)
    return np.array(tmp)

result = split_x(dataset, timesteps)
print(result)

x = result[:, :-1] 
y = result[:, -1]
print("x:\n", x, "\ny: ", y)
print("x.shape: ", x.shape, "y.shape: ", y.shape)

x = x.reshape(6, 4, 1)
```

```python
model = Sequential()                                                                      # (None, 3, 1)
model.add(LSTM(units = 64, input_shape=(4, 1), return_sequences=True, activation='relu')) # model.add(SimpleRNN(units = 64, input_length = 3, input_dim = 1)) # model.add(SimpleRNN(units = 64, input_shape=(3, 1)))
model.add(LSTM(64, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(64))
model.add(Dense(64))
model.add(Dense(16))
model.add(Dense(1))
model.summary() 

# LSTM을 두 개의 layer에서 사용할 때에는 return_sequences 파라미터를 지정해야한다. (default = False)
```

# 컴파일, 훈련

```python
model.compile(loss='mse', optimizer= 'adam')
model.fit(x, y, epochs=256)
```

# 평가, 예측

```python
loss = model.evaluate(x, y)
print(loss)

y_predict = np.array([[7, 8, 9, 10]])

result = model.predict(y_predict)

print('prediction result: ', result) 
```

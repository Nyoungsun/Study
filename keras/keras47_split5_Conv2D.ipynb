```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Conv2D, Flatten
from sklearn.model_selection import train_test_split
```

# 데이터

```python
data = np.array(range(1, 101)) 
pre_data = np.array(range(96, 106))  # [ 96  97  98  99 100 101 102 103 104 105]
print("data.shape :", data.shape)
print("pre_data.shape: ", pre_data.shape)

timesteps = 5
def split_x(dataset, timesteps):
    tmp = []
    for i in range(len(dataset) - timesteps + 1):  
        subset = dataset[i : (i + timesteps)]
        tmp.append(subset)
    return np.array(tmp)

split_data = split_x(data, timesteps) 
split_pre_data = split_x(pre_data, timesteps)
print("split_data.shape :", split_data.shape)
print("split_pre_data.shape: ", split_pre_data.shape)

# ------------------- slicing --------------------- #
data_x = split_data[:, :-1] # 마지막 열만 제거
data_y = split_data[:, -1] # 마지막 열만 추출
print("data_x.shape: ", data_x.shape, "data_y.shape: ", data_y.shape)
# print(data_x)
# print(data_y)

pre_data = split_pre_data[:, :-1] # 마지막 열만 제거
print("slicing_pre_data.shape : ", pre_data.shape)
print("slicing_pre_data:\n", pre_data)

x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, train_size=0.7, random_state=3333) # default train_size = 0.75

print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)

x_train = x_train.reshape(-1, 4, 1, 1)
x_test = x_test.reshape(-1, 4, 1, 1)
pre_data = pre_data.reshape(-1, 4, 1, 1)

print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
model = Sequential()                                                                     
model.add(Conv2D(filters = 64, kernel_size= (2, 1), input_shape=(4, 1, 1), activation='relu')) 
model.add(Conv2D(filters = 64, kernel_size= (2, 1), activation = 'relu'))
model.add(Flatten())
model.add(Dense(64, activation = 'relu'))
model.add(Dense(64))
model.add(Dense(1))
model.summary() 

# LSTM을 두 개의 layer에서 사용할 때에는 return_sequences 파라미터를 지정해야한다. (default = False)
```

# 컴파일, 훈련

```python
model.compile(loss='mse', optimizer= 'adam')
model.fit(x_train, y_train, epochs=128, batch_size=3)
```

# 평가, 예측

```python
loss = model.evaluate(x_test, y_test)
print(loss)

result = model.predict(pre_data)

print('prediction result:\n', result) 
```
